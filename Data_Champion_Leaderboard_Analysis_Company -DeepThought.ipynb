# Renaming "Columns" to avoid conflicts

```python
basic_screening.rename(columns={
    'Total': 'Score_Basic',  
    'Name': 'Name_Basic'    
}, inplace=True)

advanced_screening.rename(columns={
    'Total': 'Score_Advanced',  
    'Name (Pls enter the same name as you did in preliminary screening- we have an automation in place, so you\'d have to enter the same name)': 'Name_Advanced'
}, inplace=True)
```

# Merge datasets on the common column called 'Name'

```python
leaderboard = pd.merge(
    basic_screening,
    advanced_screening,
    left_on='Name_Basic',
    right_on='Name_Advanced',
    suffixes=('_Basic', '_Advanced'),
    how='outer'  
)
```

# Filled the missing scores with 0 for unmatched rows

```python
leaderboard['Score_Basic'] = leaderboard['Score_Basic'].fillna(0)
leaderboard['Score_Advanced'] = leaderboard['Score_Advanced'].fillna(0)
```

# Calculating scores

```python
leaderboard['Basic'] = leaderboard['Score_Basic']
leaderboard['Advanced'] = leaderboard['Score_Advanced']
leaderboard['Total'] = leaderboard['Basic'] + leaderboard['Advanced']
```

# Calculated percentage

- Updating the max_score is 30

```python
max_score = 30  
leaderboard['Percentage'] = (leaderboard['Total'] / max_score) * 100
leaderboard['Percentage'] = leaderboard['Percentage'].round(0)
```

# Ranking and Sort the Leaderboard

```python
leaderboard['Rank'] = leaderboard['Total'].rank(ascending=False, method='min')
leaderboard = leaderboard.sort_values('Rank', ascending=True)
```

# Added here the Column name or serial number as (SL No)

```python
leaderboard['SL No'] = range(1, len(leaderboard) + 1)
```

# Applying color coding based on "Total Scores"

- if score < 5  = Color = "Red"

- elif score < 10 = color = "Yellow"

- else: return color =  "Green"

```python
def color_code(score):
    if score < 5:
        return 'red'  
    elif score < 10:
        return 'yellow'  
    else:
        return 'green'  

leaderboard['Color'] = leaderboard['Total'].apply(color_code)
```

# Extracting required columns for final output

```python
leaderboard_final = leaderboard[['SL No', 'Name_Basic', 'Advanced', 'Basic', 'Total', 'Percentage', 'Color']]
```

# Saving the final "Leaderboard" to the specified output path

```python
output_path = r"C:\Users\DELL\Music\Data Analysis - Data Champion\Leaderboard_Output.xlsx"

leaderboard_final.to_excel(output_path, index=False)

print(f"Leaderboard successfully saved to {output_path}")
```
